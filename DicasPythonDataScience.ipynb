```python
#Aplicando regras e condições nas duas variveis
#a função definiu um comportamento customizado mas é possível realiz diretamente funções lambda
#Veja que existem 3 maneiras de aplica nív eis a um set de duas colunas ou mais

def rule(x,y):
    if x == 'alto' and y > 10:
        return 1
    else:
        return 0
df = pd.DataFrame({'c1':['Alto','Alto','baixo','baixo'],'c2':[0,23,17,4]})    
df['new'] = df.apply(lambda x:rule(x['c1'],x['c2']), axis = 1)
#Aplincando nível mais rapidamente que a função anterior
#df['Maximo'] = df.apply(lambda x: max(x['c1'], x['c2']), axis = 1)
#Um comando mais sintetico de aplicação de nível
df['max'] = df[['c1','c2']].max(axis = 1)
#Arredondar - é possível usar e ficar mais rapido funções do tipo round(df['c1'],0)
#porém podemois usar um apply pra essa maneira também
#df.apply(lambda x: round(x['c1'], 0), axis = 1)

df.head()
```

```python
#is null ou .sum
#elminando valores nulos das tabelas

df = pd.DataFrame({ 'id': [1,2,3], 'c1':[0,0,np.nan], 'c2': [np.nan,1,1]})
df = df[['id', 'c1', 'c2']]
df['num_nulls'] = df[['c1', 'c2']].isnull().sum(axis=1)
df.head()
```
