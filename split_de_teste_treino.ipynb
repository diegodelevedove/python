```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression as lm
```

```python
#Selecionar o conjunto de dados que vamos dividir

data=pd.read_csv('C:/Users/Diego/Desktop/pythonRead/forestfires.csv')
data.head()
```

```python
y = data.temp #definindo apenas o eixo do y como importante para a predição 
x = data.drop('temp',axis = 1) # ignorando o eixo X da predição
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2) # separando os dados - o 0.2 equivale a dizer que 20% dos dados do nosso dataset foram treinados
x_train.head()
```
